```python
#分组和排序
reviews.groupby("Pclass").count()
reviews.groupby("Pclass").mean()
reviews.groupby("Pclass").min()
reviews.groupby("Pclass").apply(lambda df : df.Age.iloc[0])
reviews.groupby(['Pclass', 'Survived']).apply(lambda df: df.loc[df.Age.argmax()])
reviews.groupby(['Pclass']).Age.agg([len, min, max])
reviews.groupby(['Pclass']).Age.count()
countries_reviewed = reviews.groupby(['Pclass', 'Survived']).Age.agg([len])
countries_reviewed.index
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len', ascending=False)
countries_reviewed.sort_values(by=['Pclass','len'], ascending=False)
countries_reviewed.sort_index()
```

```python
#数据类型和缺失数据处理
reviews.Age.dtype
reviews.Age.astype('str')
~reviews.Age.isnull()
#count不考虑null，len考虑
reviews.Age.loc[reviews.Age.isnull()].count()
len(reviews.Age.loc[reviews.Age.notnull()])
reviews.Age.fillna(0).sort_values()
reviews.Cabin.fillna('Unkown').value_counts().sort_values()
reviews.Cabin.fillna('Unkown').count()
reviews.Cabin.isnull().sum()
#重命名和多个表结合
reviews.rename(columns={'Age':'region','Cabin':'locale'})
reviews.rename_axis('wines',axis='rows')
pd.concat([reviews,stocks])
reviews.set_index('Age').join(stocks.set_index('high'),lsuffix='_CAN', rsuffix='_UK',how='right')
```

```python
#chaining
reviews.Cabin\
.replace("C85",None)\
.astype('str')\
.dropna()\
.count()

reviews.assign(
    Age=reviews.apply(lambda srs: srs.Age if pd.notnull(srs.Age) else srs.Fare, 
                        axis='columns')
)
#等价于
reviews.Age = \
    reviews.apply(lambda srs: srs.Age if pd.notnull(srs.Age) else srs.Fare, 
                        axis='columns')

reviews.assign(n=0)

def name_index(df):
    df.index.name = 'review_id'
    return df

reviews.pipe(name_index)

reviews.apply(lambda srs: srs.iloc[0], 
                        axis='rows')
```
